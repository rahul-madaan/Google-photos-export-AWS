## Rename json files which have (1) in them

```python
# write a code to rename json files which have '(1)' in .json file name

metadata_mapping = {}
for file in missing_metadata:
    if "(1)" in file:
        name, ext = os.path.splitext(file)
        name = name.rstrip(')').rstrip('(1')
        main_name, main_ext = os.path.splitext(name)
        corrected_file = f"{main_name}(1){main_ext}{ext}"
        metadata_mapping[file] = corrected_file

print(metadata_mapping)

for old_metadata, new_metadata in metadata_mapping.items():
    os.rename(folder_path+old_metadata,folder_path+new_metadata)
```
